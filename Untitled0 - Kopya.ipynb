```python
# --- KOD BAŞLANGICI ---
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import os
import zipfile

# 1. VERİ İNDİRME (Otomatik)
print("Veri indiriliyor, lütfen bekle...")
!wget --no-check-certificate \
    https://github.com/sartajbhuvaji/Brain-Tumor-Classification-DataSet/archive/refs/heads/master.zip \
    -O /content/beyin_tumoru.zip

with zipfile.ZipFile('/content/beyin_tumoru.zip', 'r') as zip_ref:
    zip_ref.extractall('/content/')

base_dir = '/content/Brain-Tumor-Classification-DataSet-master/Training'

# 2. VERİYİ HAZIRLAMA
train_datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

train_generator = train_datagen.flow_from_directory(
    base_dir, target_size=(150, 150), batch_size=32,
    class_mode='categorical', subset='training')

validation_generator = train_datagen.flow_from_directory(
    base_dir, target_size=(150, 150), batch_size=32,
    class_mode='categorical', subset='validation')

# 3. MODELİ KURMA
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)),
    MaxPooling2D(2, 2),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(4, activation='softmax')
])

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# 4. EĞİTİM (Biraz sürer)
print("Eğitim başladı...")
history = model.fit(train_generator, epochs=5, validation_data=validation_generator)

# 5. GRAFİK KAYDETME
plt.figure(figsize=(10, 5))
plt.plot(history.history['accuracy'], label='Eğitim Başarısı')
plt.plot(history.history['val_accuracy'], label='Test Başarısı')
plt.title('Beyin Tümörü Tespit Başarısı')
plt.legend()
plt.savefig('basari_grafigi.png') # Resmi kaydeder
print("TAMAMLANDI! Grafiğin hazır.")
# --- KOD BİTİŞİ ---
```
